```python
from matplotlib.pylab import plt
```

```python
losses = []
maps = []
with open("./I3D/outputs/R50_hands_obj_64x2/stdout.log") as f:
#with open("/home/soroush/Projects/slowfast/fixed_experiments/raw/stdout.log") as f:
    lines = f.readlines()
    
    for line in lines:
        if "loss" in line:
            num = line.split(" ")[-3][0:-1]
            losses.append(float(num))
            
        if "\"map\"" in line:
            num = line.split(" ")[-3][0:-1]
            maps.append(float(num))
    
```

```python
plt.plot(losses)
plt.title("loss curve")
plt.xlabel("iter")
plt.ylabel("loss")
plt.ylim(0, 0.06)
print(len(losses))
```

```python
print(maps)
plt.plot([x*5 for x in range(len(maps))], maps)
plt.title("validation mAP curve")
plt.xlabel("epoch")
plt.ylabel("mAP")
print(len(maps))
```

```python
for idx, val in enumerate(maps):
    print((idx+1)*5, val)
```
